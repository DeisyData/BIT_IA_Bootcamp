# Parte 2. Cargar las imagenes de prueba y procesarlas

*   Redimensionar la imagen 224x224
*   Convertir la imagen a un tensor
*   Normalizar la imagen



```python
!pip install tensorflow
```

```python
import numpy as np
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_test = mpimg.imread("/content/Steak2.jpeg")
plt.imshow(img_test)
print(f'Tamaño original', img_test.shape)
img_test = tf.image.resize(img_test, size=[224,224])
img_test = tf.cast(img_test, dtype=tf.float32)
img_test = img_test/255.
#img_test = tf.expand_dims(img_test, axis=0)
print(f'Tamaño ajustado', img_test.shape)
img_test_array = img_to_array(img_test)
img_test_array = np.expand_dims(img_test_array, axis = 0)
```

# Parte 3. Predicciones
