```python
import numpy as np
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_test = mpimg.imread("/content/steak_3.jpeg") # cargar la imagen
plt.imshow(img_test) # Imprimir la imagen
print(f'Tamaño original', img_test.shape)  # verificacion del tamaño incial
img_test = tf.image.resize(img_test, size=[224,224]) # Ajusta la imagen al tamaño 224x224
img_test = tf.cast(img_test, dtype=tf.float32) # Ajuste dek tensor
img_test = img_test/255. # Normalizan las imagenes
#img_test = tf.expand_dims(img_test, axis=0)
print(f'Tamaño ajustado', img_test.shape)
img_test_array = img_to_array(img_test)
img_test_process = np.expand_dims(img_test_array, axis = 0)
```

# Parte 3. Predicciones

```python
prediccion1 = model_5.predict(img_test_process)
pred1 = prediccion1[0][0]
pred_percent1 = pred1*100

if pred1 < 0.5:
  print(f'Clase 0 : {pred1}' )
  pred_percent1 = 100 - pred_percent1
  print(f'Con probabilidad de ser clase 0 (PIZZA) es de  {pred_percent1} %')
else:
  print(f'Clase 1: {pred1}')
  print(f'Con probabilidad de ser clase 1 (STEAK) es de {pred_percent1} %')
```

```python
prediccion2 = model_6.predict(img_test_process)
pred2 = prediccion2[0][0]
pred_percent2 = pred2*100

if pred2 < 0.5:
  print(f'Clase 0 : {pred2}, con probabilidad ' )
  pred_percent2 = 100 - pred_percent2
  print(f'Con probabilidad de ser clase 0 (PIZZA) es de  {pred_percent2} %')
else:
  print(f'Clase 1: {pred2}')
  print(f'Con probabilidad de ser clase 1 (STEAK) es de {pred_percent2} %')
```
