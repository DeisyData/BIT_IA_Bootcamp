```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ruta del archivo que deseas cargar
file_path = '/content/auto-mpg.data-original'

column_names = ["mpg", "cylinders", "displacement", "horsepower", "weight",
                "acceleration", "model year", "origin", "car name"]
# Leer el archivo usando delimitador de espacios en blanco
df = pd.read_csv(file_path, delim_whitespace=True, header=None, names=column_names)

df.head(5)
```
