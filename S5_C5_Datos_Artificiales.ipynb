```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, Y = make_classification(
    n_samples = 500, n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1
) # Se genera un conjunto de datos para clasificación
plt.scatter(X[:, 0], X[:, 1], marker="o", c=Y, s=25, edgecolor="k") # Hacemos el gráfico de dispersión
```
