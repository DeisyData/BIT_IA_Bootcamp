```python
import tensorflow as tf
```

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img_test = mpimg.imread("/content/steaak2.jpg")
plt.imshow(img_test)
img_test.shape
```

```python
img_test = tf.image.resize(img_test, size=[224,224])
img_test = tf.cast(img_test, dtype=tf.float32)
img_test = img_test/255.
#img_test = tf.expand_dims(img_test, axis=0)
img_test.shape
```

```python
import numpy as np
```

```python
img_test_array = img_to_array(img_test)
img_test_array = np.expand_dims(img_test_array, axis = 0)
```

# Parte 3. Predicciones

```python
prediccion = model_5.predict(img_test_array)
```
